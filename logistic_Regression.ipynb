```python
import pandas as pd
import numpy as np
import math
```

```python
DataPath = ('diabetes.csv')
data = pd.read_csv(DataPath)
data.head()
```

```python
data.sample(frac=1).reset_index()
```

```python
features = ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
X1 = data[features]
y = data.Outcome
# print(X1)
# print(y)
```

```python
X1=(X1-X1.min(axis=0))/(np.ptp(X1,axis=0))

X1
```

```python
X1.insert(0,'1s',1)
X1
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (X1,y,test_size = 0.20, random_state = 0)
```

```python
lr=0.01
loss_list=[]
total_loss=0.0
theta_old=np.random.rand(9,1)
theta_new=None
theta_old

```

```python
def stable_sigmoid(x):
    sig = np.where(x < 0, np.exp(x) / (1 + np.exp(x)), 1 / (1 + np.exp(-x)))
    return sig
for e in range(100):
  Z=np.matmul(x_train,theta_old)
  h=stable_sigmoid(Z)
  j_vect=(-y_train*(np.log2(h[e])))-((1-y_train)*np.log2(1-h[e]))
  j=np.mean(j_vect)
  total_loss+=j
  loss_list.append(total_loss)
  grad=np.matmul(x_train.T,(h[e]-y_train))
  theta_new=theta_old-(lr*grad)
  theta_old=theta_new
  # print(theta_new)





```

```python
from sklearn import metrics
def stable_sigmoid(x):
    sig = np.where(x < 0, np.exp(x) / (1 + np.exp(x)), 1 / (1 + np.exp(-x)))
    return sig
for e in range(100):
  Z=np.matmul(x_test,theta_old)
  h=stable_sigmoid(Z)
  h[h>=0.5]=1
  h[h<0.5]=0
  y_test_predicted=h
print("Accuracy : ", metrics.accuracy_score(y_test, y_test_predicted))
```
